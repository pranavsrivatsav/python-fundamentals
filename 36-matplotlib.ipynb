#### Basic Line Plot
- To create a basic line plot we need the coordinates. The x axis value of the coordinates in a separate array and the y value of the coordinates in a separate one.
- And then pass these coordinates to the plot function of pyplot (imported as plt)


```python
xvalues = [1,2,3]
yvalues=[2,5,7]
plt.plot(xvalues, yvalues) #this will also return the line chart, but along with the reference
plt.title("Simple line plot") # set title
plt.xlabel("x Axis label") # set x axis label
plt.ylabel("y Axis label") # set y axis label
plt.title(label="plot label",fontdict={"font": "comic sans ms", "size": 20}) # set title along with font props
plt.xlabel(xlabel="x label",fontdict={"font": "comic sans ms", "size": 10}) # set xlabel along with font props
plt.ylabel(ylabel="y label",fontdict={"font": "comic sans ms", "size": 10}) # set ylabel along with font props
# plt.yticks(ticks=[3,6,9])
plt.show() #use show function to show only the line chart


#from hereon we will call this a plot
```

#### Bar Chart

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [15, 25, 10, 30]
plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

#### Histogram

- A histogram when provided an array, will provide us a graph indicating a count of the occurence of values in the array.
- Bins: The range of the array is covered in the x axis using bins, we can specify the number of bins.

```python
import pandas as pd

# We will create a dataframe from the olympians bios csv
bios = pd.read_csv("35-pandas/data/bios.csv")

print(bios.head())

# Plot a histogram for the height of the athletes
ax = bios['height_cm'].plot(kind='hist', bins=20, title='Histogram of Values', edgecolor='black')

# Add labels
ax.set_xlabel('Height (cm)')
ax.set_ylabel('No of athletes')

# Show the plot
plt.show()
```

#### Scatter plot
- It takes values of x axis and y axis as separate arrays and marks a point on the plot

```python
# let us put a scatter plot of height and weight of the olympians
ax = bios.plot(kind='scatter', x='weight_kg', y='height_cm', title='Scatter Plot of height vs weight', color="red", s=20)
#s - size of points
#color - color of points
# Add labels
ax.set_xlabel('weight(kg)')
ax.set_ylabel('height(cm)')

# Show the plot
plt.show()
```
